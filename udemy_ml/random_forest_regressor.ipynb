```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('Position_Salaries.csv')
X = df.iloc[:, 1:-1]
y = df.iloc[:, -1]
```

# random forest 
- pick random K data points
- build decision tree associated with these K data points
- choose the number of N trees you want to build and repeat first two steps
- for a new data point, make each N tree predict value of y, and take average and that is your predicted result

relatively robust to outliers

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=300, random_state=0)
regressor.fit(X, y)
```

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, regressor.predict(X), color='blue')
plt.title('random forest regression truth or bluff')
plt.xlabel('level')
plt.ylabel('salary')
```

```python
X_grid = np.arange(np.amin(X), np.amax(X), .01)
X_grid = X_grid.reshape((-1, 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, regressor.predict(X_grid), color='blue')
plt.title('random forest regression truth or bluff')
plt.xlabel('level')
plt.ylabel('salary')
```
